```python
# import pygame's library, pygame contains many different modules
import pygame
# import system's exit function to be used
from sys import exit

# import python's numpy library, numpy contins plenty of mathematical tools)
import numpy as np
# import python's pandas library, pandas contains the tools for effective data management)
import pandas as pd 

# check the encoding of the csv file
with open("NTU map_dataset.csv") as f: 
    print (f)
    
# read database in csv file and assign to variable name of dataset
data_set = pd.read_csv('NTU map_dataset.csv', encoding = "cp1252")
# to display first 10 rows of data inside csv file
data_set.head(10)

# initialize all imported pygame modules
pygame.init()
screen = pygame.display.set_mode((700,870), 0, 32)

pygame.display.set_caption("NTU map")
background = pygame.image.load("NTU map.png").convert()

def distance():
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()   
                exit()
    
            if event.type == pygame.MOUSEBUTTONDOWN:
                screen.blit(background, (0,0))
                
                # to provide coordinate position based off mouse click  
                x1, y1 = pygame.mouse.get_pos()
                print("the coordinate is: ", x1, y1)
    
                pygame.display.update()
                
                # calculating the distance between current location and possible canteens
                distance_list=[]
                for i in range(len(data_set)):
                      distance_list.append(np.abs(int(data_set['x'][i])-x1)+np.abs(int(data_set['y'][i]-y1)))
                # data in database is aligned in a table format with rows and columns
                df = pd.DataFrame(data_set)
                print (df)

                df['Distance'] = distance_list
                # sorting by distances in ascending order
                df.sort_values(by=['Distance'], inplace=True, ascending=True)
                print(df)
                
                # select the nearest canteen by position based off 'Distance' value
                closest_canteen = df.iloc[0,0]
                print(closest_canteen)
                print(sorted(distance_list))
                print("The nearest canteen is",(closest_canteen), "which is", min(distance_list), "units away.")

# to define a reuseable code to provide the food prices of the canteens
def foodprice():
    fp = int(input("Enter your average price:"))
    while fp < 6:
        if fp == 3:
            print("Canteen 9, Canteen 13, Canteen 14 and South Spine Food Court")
            break
        elif fp == 4:
            print("Canteen 2, Canteen 11, Canteen 16 and North Spine Food Court")
            break
        elif fp == 5:
            print("Canteen 1 and Quad Cafe")
            break
        else:
            print("Price is too low, please try again.")
            break
    else:
        print("Price is too high, please try again.")
        
# to define a reuseable code to provide the operating hours of the canteens
def operatinghours():
    opshrs = int(input("Enter your hour of visit (0-23):"))
    if opshrs > 21:
        print("Timing is too late, no canteens are still open.")
    elif opshrs < 7:
        print("Timing is too early, no canteens are open yet.")
    else:
        print("All of the canteens are open.")
        
# to define a reuseable code to provide the number of stalls at the canteens
def numstalls():
    nst = int(input("Enter the number of stalls:"))
    while nst < 20:
        if nst <= 8:
            print("Canteen 11, Canteen 13, Canteen 14, Canteen 16 and Quad Cafe has about stalls.")
            break
        elif nst <= 10:
            print ("Canteen 1, Canteen 2, Canteen 9 has about 10 stalls.")
            break
        elif nst > 10:
            print("North Spine Food Court and South Spine Food Court have more than 10 stalls.")
            break
    else:
        print("Number of stalls is too high. There are no canteens with that many stalls, please try again.")

# to define a reuseable code to provide the seating capacity of the canteens        
def seatingcapacity():
    seats = int(input("Enter the seating capacity:"))
    if seats < 211:
        print("Canteen 11 and 13 has a seating capactiy of 210.")
        
    elif seats > 210 and seats < 271:
        print("Canteen 14 has a seating capacity of 270.")
      
    elif seats > 270 and seats < 293:
        print("Canteen 9 has a seating capacity of 292.")
        
    elif seats > 292 and seats < 305:
        print("Canteen 16 has a seating capacity of 304.")
     
    elif seats > 304 and seats < 311:
        print("Canteen 1 has a seating capacity of 310.")
       
    elif seats > 310 and seats < 447:
        print("Canteen 2 has a seating capacity of 446.")
     
    elif seats > 446 and seats < 501:
        print("Quad cafe has a seating capacity of 500.")
      
    elif seats > 500 and seats < 801:
        print("North Spine Food Court has a seating capacity of 800.")
        
    elif seats > 800 and seats < 1051:
        print("South Spine Food Court has a seating capacity of 1050.")
        
    elif seats > 1050:
        print("There is no canteen with a seating capacity of more than 1050. Please try a lower number.")
        
    else:
        print("There is no canteen with a seating capacity of less than 210, please try again.")

# to define a reuseable code to provide the dietary options at the canteens 
def dietaryrequirement():
    dietreq = str(input("What dietary requirement do you have? (Halal, Vegetarian, None):"))
    while dietreq == "Halal" or "Vegetarian" or "None":
        if dietreq == "Halal":
            print("Canteen 2, Canteen 9, Canteen 11, Canteen 13, Canteen 14, Canteen 16, Quad Cafe, North Spine Food Court and South Spine Food Court have Halal food stalls.")
            break
        
        elif dietreq == "Vegetarian":
            print("Quad Cafe, North Spine Food Court and South Spine Food Court have Vegetarian food stall.")
            break
        
        elif dietreq == "None":
            print("You can go to any canteen to enjoy the food.")
            break
        
        else:
            print("Please provide with either Halal, Vegetarian or None.")
            break
        
    else:
        print("Please provide with either Halal, Vegetarian or None.")

# to enquire user to select an option for their criteria
def showmsg():
    print()
    print("Hello, please select an option:")
    print("1.Distance to the Canteen")
    print("2.Average price of food at the Canteen")
    print("3.Operating Hours of the Canteen")
    print("4.Number of Stalls at the Canteen")
    print("5.Seating Capacity at the Canteen")
    print("6.Dietary options at the Canteen")
    print("7.Exit:")
    
# input values for the user to select their criteria
Distance = 1
Price = 2
OperatingHours = 3
NumberofStalls = 4
SeatingCapacity = 5
DietRequirement = 6
Exit = 7

option = Distance
while option !=Exit:
    showmsg()
    option = int(input("Option:"))
    if option == Distance:
        data_set = pd.read_csv('NTU map_dataset.csv')
        # dropping unimportant columns of Avg price, Operation hours, No. of stalls, Seating capacity, Dietary requirement
        data_set.drop(columns=['Avg price','Operation hours','No. of stalls','Seating capacity','Dietary requirement'], inplace = True)
        print (data_set)
        i = distance()
        
    elif option == Price:
        data_set = pd.read_csv('NTU map_dataset.csv')
        print (data_set)
        # dropping unimportant columns of x, y, Operation hours, No. of stalls, Seating capacity, Dietary requirement
        data_set.drop(columns=['x', 'y','Operation hours','No. of stalls','Seating capacity','Dietary requirement'], inplace = True)
        print(data_set)
        fp = foodprice()
        
    elif option == OperatingHours:
        data_set = pd.read_csv('NTU map_dataset.csv')
        print (data_set)
        # dropping unimportant columns of x, y, Avg price, No. of stalls, Seating capacity, Dietary requirement
        data_set.drop(columns=['x', 'y','Avg price','No. of stalls','Seating capacity','Dietary requirement'], inplace = True)
        print(data_set)
        opshrs = operatinghours()
    
    elif option == NumberofStalls:
        data_set = pd.read_csv('NTU map_dataset.csv')
        # dropping unimportant columns of x, y, Avg price, Operation hours, Seating capacity, Dietary requirement
        data_set.drop(columns=['x', 'y','Avg price','Operation hours','Seating capacity','Dietary requirement'], inplace = True)
        print(data_set)
        nst = numstalls()
    
    elif option == SeatingCapacity:
        data_set = pd.read_csv('NTU map_dataset.csv')
        # dropping unimportant columns of x, y, Avg price, Operation hours, No. of stalls, Dietary requirement
        data_set.drop(columns=['x', 'y','Avg price','Operation hours','No. of stalls','Dietary requirement'], inplace = True)
        print(data_set)
        Seats = seatingcapacity()
        
    elif option == DietRequirement:
        data_set = pd.read_csv('NTU map_dataset.csv')
        # dropping unimportant columns of x, y, Avg price, Operation hours, No. of stalls, Seating capacity, 
        data_set.drop(columns=['x', 'y','Avg price','Operation hours','No. of stalls','Seating capacity'], inplace = True)
        print(data_set)
        food = dietaryrequirement()
        
    else:
        option == Exit
        exit()
```
