```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import time
```

```python
start_time = time.time()
dataset = pd.DataFrame()

oly_years=[2008,2012,2016]
for year in oly_years:
    link = 'https://en.wikipedia.org/wiki/{}_Summer_Olympics_medal_table'.format(year)
    html_data = pd.read_html(link)[1]
    html_data['Year'] = year
    dataset = dataset.append(html_data)
    
    #dataseth = pd.DataFrame()
    #html_datah = pd.read_html(link)[1].head()
    #html_datah['Year'] = year
    #dataseth = dataseth.append(html_data)
    
print(dataset)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()

unique_countries = dataset['NOC'].unique()
for country in unique_countries[:-1]:  # the last element is the total number of medals for all countries
    bronze,silver,gold,total = 0,0,0,0
    for row,column in dataset.iterrows():
        if column['NOC'] == country:            
            gold += column['Gold']
            silver += column['Silver']
            bronze += column['Bronze']
            total += column['Total']
    print('The number of golds {0} won is:\ngold = {1},\nsilver = {2},\nbronze = {3}'.format(country,gold,silver,bronze))
    
print(time.time()-start_time)
    
        
# for row, column in dataset.iterrows():
#     if column['NOC'] == 'United States\xa0(USA)':
#         gold += column['Gold']
#         silver += column['Silver']
#         bronze += column['Bronze']
#         total += column['Total']
# print('The number of golds the United States won is:\ngold = {0},\nsilver = {1},\nbronze = {2}'.format(gold,silver,bronze))
    
```
